```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([12,13,56,23,41,28,51])
y = np.array([51,63,37,88,33,72,81])
mean_x = np.mean(x)
print(mean_x)
```

```python
median_x = np.median(x)
print(median_x)
```

```python
mode_x =pd.Series(x).mode().values
print(mode_x)
```

```python
variance_x = np.var(x)
print(variance_x)
```

```python
std_dev_x = np.std(x)
print(std_dev_x)
```

```python
range_x = np.max(x)-np.min(x)
print(range_x)
```

```python
iqr_x = np.percentile(x, 75) - np.percentile(x, 25)
print(iqr_x)
skewness_x = (3 * (mean_x - median_x)) / std_dev_x
print(skewness_x)
kurtosis_x = np.mean((x - mean_x)**4) / (std_dev_x**4) - 3
print(kurtosis_x)
```
